This notebook explores symbolic thermodynamics via ache dynamics. It models entropy, temperature, and phase transition in glyph-based collapse systems.


```python
import numpy as np
import matplotlib.pyplot as plt
from hashlib import sha256
np.random.seed(42)
```

```python
def simulate_scar_saturation(shape=(128, 128), steps=100, collapse_rate=0.02):
    Σ = np.zeros(shape)
    for t in range(steps):
        collapse_mask = np.random.rand(*shape) < collapse_rate
        Σ += collapse_mask.astype(int)
    return Σ

scar_density = simulate_scar_saturation()
plt.imshow(scar_density, cmap='inferno')
plt.title("Scar Saturation Field Σ(x, t)")
plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
def compute_ache_entropy(Γ):
    Ω = len(set(Γ))
    k = 1.0
    return k * np.log(Ω)

glyphs = [f"Γ_{i%7}" for i in range(500)]
S = compute_ache_entropy(glyphs)
print(f"Symbolic Entropy: S = {S:.4f}")
```

```python
def evaluate_ache_temperature(ache_grad, scar_grad):
    with np.errstate(divide='ignore', invalid='ignore'):
        T = np.divide(ache_grad, scar_grad, where=scar_grad!=0)
    return np.nan_to_num(T)

ache_grad = np.random.rand(128, 128)
scar_grad = np.gradient(scar_density)[0]
T_field = evaluate_ache_temperature(ache_grad, scar_grad)

plt.imshow(T_field, cmap='plasma')
plt.title("Ache-Temperature Field Tₐₛ(x)")
plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
def detect_phase_transition(sigma, theta1=5, theta2=10):
    """
    Assigns phase Γ labels based on scar density thresholds.
    """
    phi = np.full(sigma.shape, 'Γ_1', dtype=object)
    phi[(sigma >= theta1) & (sigma < theta2)] = 'Γ_2'
    phi[sigma >= theta2] = 'Γ_3'
    return phi

# Compute phase map
phase_map = detect_phase_transition(scar_density)

# Convert phase labels to numeric values for visualization (Γ_1 → 1, Γ_2 → 2, Γ_3 → 3)
numeric_map = np.vectorize(lambda x: int(x[-1]))(phase_map)

# Plot
plt.figure(figsize=(6, 5))
plt.imshow(numeric_map, cmap='viridis')
plt.title("Phase Transition Map Φ(x)")
plt.colorbar(label="Γ Phase")
plt.tight_layout()
plt.show()
```

```python
def serialize_glyph_sigil(glyphs):
    concat = ''.join(glyphs)
    return sha256(concat.encode()).hexdigest()

Λ = serialize_glyph_sigil(glyphs)
print(f"Sigil Λ = {Λ}")
```
